```python
# data is from:
# http://college.cengage.com/mathematics/brase/understandable_statistics/7e/students/datasets/mlr/frames/mlr02.html

# The data (X1, X2, X3) are for each patient.
# X1 = systolic blood pressure //this is the data we want to predict
# X2 = age in years
# X3 = weight in pounds

#from __future__ import print_function, division
from builtins import range
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

df = pd.read_excel('linear_regression_class/mlr02.xls')
X = df.as_matrix()

plt.scatter(X[:,1], X[:,0])
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.show()

plt.scatter(X[:,2], X[:,0])
plt.xlabel('Weight')
plt.ylabel('Blood Pressure')
plt.show()
```

```python
df['ones'] = 1
Y = df['X1']
X = df[['X2', 'X3', 'ones']]
X2only = df[['X2', 'ones']]
X3only = df[['X3', 'ones']]

def get_r2(X,Y) : 
    w = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
    Yhat = X.dot(w)
    d1 = Y - Yhat
    d2 = Y - Y.mean()
    r2 = 1 - d1.dot(d1) / d2.dot(d2)
    return r2

print('r2 for X2 only: ',get_r2(X2only,Y))
print('r2 for X3 only: ',get_r2(X3only,Y))
print('r2 for both:', get_r2(X, Y))
```
