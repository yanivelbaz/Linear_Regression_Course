```python
# shows how linear regression analysis can be applied to polynomial data

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

#load the data
X = []
Y = []
for line in open('linear_regression_class/data_poly.csv') : 
    x,y = line.split(',')
    x = float(x)
    X.append([1,x,x*x]) # add the bias term x0 = 1
    # our model is therefore y_hat = w0 + w1 * x + w2 * x**2
    Y.append(float(y))
    
X = np.array(X)
Y = np.array(Y)

plt.scatter(X[:,1], Y)
plt.title("The data we're trying to fit")
plt.show()
```

```python
# calculate weights
w = np.linalg.solve(np.dot(X.T,X), np.dot(X.T,Y))
Yhat = np.dot(X, w)

# === FIRST OPTION TO PLOT THE PREDICTED LINE ===
plt.scatter(X[:,1], Y)
plt.plot(sorted(X[:,1]),sorted(Yhat))
plt.show()
```

```python
# === SECOND OPTION TO PLOT THE PREDICTED LINE ===
# to plot our quadratic model predictions, let's
# create a line of x's and calculate the predicted y's
plt.scatter(X[:,1], Y)
x_line = np.linspace(X[:,1].min(), X[:,1].max())
y_line = w[0] + w[1] * x_line + w[2] * x_line**2
plt.plot(x_line, y_line)
plt.title("Our fitted quadratic")
plt.show()
```

```python
# compute the r-square 
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print('r-squared: ', r2)
```
