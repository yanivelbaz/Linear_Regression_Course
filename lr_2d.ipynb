```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
%matplotlib inline

#load the data
X = []
Y = []
for line in open('linear_regression_class/data_2d.csv') : 
    x1, x2, y = line.split(',')
    X.append([float(x1), float(x2), 1]) # 1 is the added bias term
    Y.append(float(y))
    
#turn X and Y into numpy arrays
X = np.array(X)
Y = np.array(Y)

# let's plot to see the data
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,0], X[:,1], Y)

plt.show()


```

```python
# calculate the weights of the model
# apply the equations we learned to calculate a and b
# numpy has a special method for solving Ax = b
# so we don't use x = inv(A)*b
# note: the * operator does element-by-element multiplication in numpy
#       np.dot() does what we expect for matrix multiplication

w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat = np.dot(X, w)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,0], X[:,1], Y)

plt.plot(sorted(X[:,0]), sorted(X[:,1]), sorted(Yhat))
plt.show()

#Xmesh, Ymesh = np.meshgrid(X[:,0], X[:,1])
#Zmesh = w[0]*Xmesh + w[1]*Ymesh + w[2]
#surf = ax.plot_surface(Xmesh, Ymesh, Zmesh)


# compute the r-square 
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print('r-squared: ', r2)
```
